```python
import lmdb
import pickle as pkl

def read_protein_seq(fn):
    with open(fn,'r') as f:
        line= eval(f.readlines()[0])
        return line
def convert_format(fn):
    data_list=read_protein_seq('%s.json'%fn)
    map_size = (1024 * 15) * (2 ** 20) # 15G, change accordingly
    wrtEnv = lmdb.open('%s.lmdb'%fn,map_size=map_size)
    with wrtEnv.begin(write=True) as txn:
        for i, entry in enumerate(data_list): # data_list contains all dictionaries in the above format
            txn.put(str(i).encode(), pkl.dumps(entry))
        txn.put(b'num_examples', pkl.dumps(i+1))
    wrtEnv.close()
for fn in ['/home/grads/z/zshuying/Documents/shuying/ppi_mutation/data/single_protein_seq/batch150_%d'%i for i in range(75)]:
    convert_format(fn)
```
